```python
%matplotlib inline 
from matplotlib import pyplot as plt
from matplotlib_venn import venn3, venn3_circles
from matplotlib import font_manager, rc



body_font = {
    "family" : "sans",
    "sans-serif" : "Gotham",
    "weight" : "light",
    "size" : 9
}

# Set default font 
rc('font', **body_font)
rc('text', usetex=False)
# Heading font
heading_font = font_manager.FontProperties(weight="medium", size=10)



def format_set(s):
    return "\n\n".join(s)

languages = set(["English\n(native speaker)", "French\n(fluent)", "HTML", "CSS", "Latex", "Javascript\n(D3.js)", "Java", "Python","R", "SQL", "XML"])
analysis = set(["Javascript\n(D3.js)", "Java", "Python","R", "SQL", "XML", "Probability", "Statistics", "Combinatorics", "Machine learning", "Natural language\nprocessing", "Gephi", "Weka", "Open Refine", "Tableau", "Excel"])
office = set(["Gephi", "Weka", "Open Refine", "Tableau", "Excel", "InDesign", "Windows", "Linux", "Latex", "HTML", "CSS", "Java", "Python", "Javascript\n(D3.js)"])


fig, ax = plt.subplots(figsize=(12, 12))

v = venn3([languages, analysis, office], ["LANGUAGES", "ANALYSIS", "OFFICE"])

for label in ['A', 'B', 'C']: 
    v.get_label_by_id(label).set_fontproperties(heading_font)
    
# v.get_patch_by_id('001').set_alpha(0.5)
# v.get_patch_by_id('001').set_color('gray')
v.get_label_by_id('001').set_text(format_set(office - languages - analysis))
v.get_label_by_id('010').set_text(format_set(analysis - languages - office))
v.get_label_by_id('011').set_text(format_set(analysis & office - languages))
v.get_label_by_id('100').set_text(format_set(languages - analysis - office))
v.get_label_by_id('101').set_text(format_set(languages &  office - analysis))
v.get_label_by_id('110').set_text(format_set(languages & analysis - office))
v.get_label_by_id('111').set_text(format_set(languages & analysis & office))


plt.show()
```

```python
import pyphen
import re

text = work["description"][0]
nchars = 20

def wrap_text(text, nchars, give=0):
    dic = pyphen.Pyphen(lang='en_EN')
    words = text.split(' ')
    wrapped_string = ""
    line = ""

    while words:
        word = words[0]
        linelength = len(line)
        if linelength + len(word) < nchars:
            if len(words) == 1:
                wrapped_string += line + word
            else:
                line += word + " "
            words = words[1:]
        else: 
            wrap = dic.wrap(word, nchars- linelength)
            if wrap and len(wrap) > give and not re.match(r'[eE]?[sS][.,\/#!$%\^&\*;:{}=\-_`~()]?', wrap[1]): # Avoid hyphenating on plural s and es
                newword = wrap[0].replace("--", "-")
                line += newword + "\n"
                if len(words) > 1:
                    words = [wrap[1]] + words[1:]
                else:
                    words = [wrap[1]]
                wrapped_string += line
                line = ""
            else: 
                line += "\n"
                wrapped_string += line
                line = ""
                
    return wrapped_string
wrap_text(text, nchars)
```
